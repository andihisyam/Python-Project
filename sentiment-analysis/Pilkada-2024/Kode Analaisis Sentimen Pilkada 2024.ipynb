```python
# Load NRC to memory
positive = []
negative = []

for i in range(1, len(dictionary)):
    kata = dictionary.iloc[i, 1]
    if dictionary.iloc[i, 6] == 1:
        positive.append(kata)
    if dictionary.iloc[i, 7] == 1:
        negative.append(kata)

processed_comments = dataset['CleanReview']
datasetangkapos = []

for kalimat in processed_comments:
    baris = []
    baris.append(kalimat)

    f_sentimenpositif = 0
    f_sentimennegatif = 0

    for teks in kalimat.split():
        for j in range(len(positive)):
            if teks == positive[j]:
                f_sentimenpositif += 1
        for j in range(len(negative)):
            if teks == negative[j]:
                f_sentimennegatif += 1

    # Menambahkan aturan jika terdapat kata "kontol", maka langsung diklasifikasikan sebagai sentimen negatif
    if "kontol" in kalimat.lower():
        label_sentimen = 'Negatif'
    # Menentukan sentimen berdasarkan perbandingan jumlah kata positif dan negatif
    elif f_sentimenpositif > f_sentimennegatif:
        label_sentimen = 'Positif'
    elif f_sentimenpositif < f_sentimennegatif:
        label_sentimen = 'Negatif'
    else:
        label_sentimen = 'Netral'

    baris.append(label_sentimen)
    datasetangkapos.append(baris)

# Membuat DataFrame dari dataset yang telah diberi label
dataset_labelled = pd.DataFrame(datasetangkapos, columns=['Review', 'Sentimen'])
```

```python
def predict_emotion(review):
    # Cek apakah kata 'kontol' ada dalam review
    if 'kontol' in review:
        return 'Anger'

    # Inisialisasi skor emosi
    emotion_scores = {
        'Anger': 0,
        'Anticipation': 0,
        'Disgust': 0,
        'Fear': 0,
        'Joy': 0,
        'Sadness': 0,
        'Surprise': 0,
        'Trust': 0
    }

    # Iterasi melalui setiap kata dalam review
    for word in review.split():
        # Periksa apakah kata tersebut ada dalam kamus NRC
        if word in dictionary['Indonesian (id)'].values:
            # Periksa label emosi untuk kata tersebut dalam bahasa Indonesia
            for emotion in emotion_scores.keys():
                if dictionary.loc[dictionary['Indonesian (id)'] == word, emotion].values[0] == 1:
                    emotion_scores[emotion] += 1
        elif word in dictionary['Javanese (jw)'].values:
            # Periksa label emosi untuk kata tersebut dalam bahasa Jawa
            for emotion in emotion_scores.keys():
                if dictionary.loc[dictionary['Javanese (jw)'] == word, emotion].values[0] == 1:
                    emotion_scores[emotion] += 1

    # Tentukan emosi dominan berdasarkan skor tertinggi
    dominant_emotion = max(emotion_scores, key=emotion_scores.get)

    # Jika tidak ada kata yang ditemukan, default prediksi ke Netral
    if all(value == 0 for value in emotion_scores.values()):
        dominant_emotion = 'None'

    return dominant_emotion
```

```python
import re
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Ekstrak mention dari full_text
def extract_mentions(text):
    mentions = re.findall(r'@([A-Za-z0-9_]+)', text)
    return mentions

# List untuk edges
edges = []

# Loop melalui setiap baris di DataFrame (gabungan_ori)
for index, row in gabungan_cleaned.iterrows():
    source = row['username']
    mentions = extract_mentions(row['full_text'])
    for target in mentions:
        edges.append((source, target))

# Buat graf menggunakan NetworkX
G = nx.DiGraph()  # Menggunakan graf terarah (directed graph)
G.add_edges_from(edges)

# Menghitung Indegree
in_degree = dict(G.in_degree())

# Sorting nodes berdasarkan Indegree dari yang terbesar ke terkecil
sorted_nodes = sorted(in_degree, key=lambda x: in_degree[x], reverse=True)

# Visualisasi Graf dengan menyesuaikan warna berdasarkan Indegree
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)  # Layout untuk node posisi

# Menentukan ukuran dan warna node berdasarkan Indegree
node_size = []
node_labels = {}
for node in G.nodes():
    if node in sorted_nodes[:10]:  # 10 node teratas
        node_size.append(3000 * in_degree[node])  # Ukuran besar untuk 10 teratas
        node_labels[node] = node  # Menambahkan label untuk 10 teratas
    else:
        node_size.append(300 * in_degree[node])  # Ukuran kecil untuk sisanya
        node_labels[node] = ''  # Label kosong untuk sisanya

# Menggambar graf dengan menyesuaikan ukuran node
nx.draw(G, pos, with_labels=False, node_size=node_size, node_color="skyblue", font_size=10, font_weight="bold", arrows=True)

# Menambahkan label dengan ukuran yang sesuai
for node, (x, y) in pos.items():
    if node in sorted_nodes[:10]:  # 10 node teratas
        plt.text(x, y, node, fontsize=12, ha='center', va='center', fontweight='bold')
    else:
        if in_degree[node] > 0:  # Hanya menampilkan label untuk node dengan indegree > 0
            plt.text(x, y, node, fontsize=6, ha='center', va='center')

# Menambahkan judul
plt.title("Analisis Tokoh Penting")

# Menampilkan graf
plt.show()
```

```python
# Ekstrak mention dari full_text
def extract_mentions(text):
    mentions = re.findall(r'@([A-Za-z0-9_]+)', text)
    return mentions

merged_df = pd.merge(dataset, dataset_labelled, left_on='CleanReview', right_on='Review', how='inner')

# Fungsi untuk membuat graf dan menentukan tokoh berpengaruh berdasarkan kelas
def find_influential_accounts(df, attribute):
    results = {}
    categories = df[attribute].unique()

    for category in categories:
        sub_df = df[df[attribute] == category]
        edges = []
        for index, row in sub_df.iterrows():
            source = row['username']
            mentions = extract_mentions(row['full_text'])
            for target in mentions:
                edges.append((source, target))

        G = nx.DiGraph()
        G.add_edges_from(edges)

        in_degree = dict(G.in_degree())
        sorted_influential = sorted(in_degree, key=in_degree.get, reverse=True)[:10]
        influential_dict = {node: in_degree[node] for node in sorted_influential}
        results[category] = influential_dict

    return results

# Analisis Sentimen
influential_accounts_sentiment = find_influential_accounts(merged_df, 'Sentimen')

# Analisis Emosi
influential_accounts_emotion = find_influential_accounts(merged_df, 'Emosi')

# Output Analisis Sentimen
print("Analisis Sentimen")
for sentiment, accounts in influential_accounts_sentiment.items():
    print(f"\nSentimen: {sentiment}")
    for account, degree in accounts.items():
        print(f"{account}: {degree}")

# Output Analisis Emosi
print("\nAnalisis Emosi")
for emotion, accounts in influential_accounts_emotion.items():
    print(f"\nEmosi: {emotion}")
    for account, degree in accounts.items():
        print(f"{account}: {degree}")
```
