```python
import cv2
from ultralytics import YOLO
import time

# Load model YOLOv11 custom
model = YOLO('best5.pt')

# Buka webcam
cap = cv2.VideoCapture(0)

# Set resolusi lebih kecil untuk mempercepat
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

if not cap.isOpened():
    print("Error: Tidak bisa membuka webcam")
    exit()

prev_time = 0

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Tidak bisa membaca frame dari webcam")
        break

    # Hitung FPS sederhana
    curr_time = time.time()
    fps = 1 / (curr_time - prev_time) if prev_time else 0
    prev_time = curr_time

    # Jalankan inferensi deteksi jerawat
    results = model(frame)

    # Gambar hasil deteksi ke frame
    annotated_frame = results[0].plot()


    # Tampilkan frame dengan hasil deteksi
    cv2.imshow('Deteksi Jerawat Real-Time', annotated_frame)

    # Tekan 'q' untuk keluar
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
