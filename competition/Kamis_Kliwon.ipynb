# Pre-Processing

## Data Train

### Mengosongkan inputan selain angka

```python
# Melihat nilai unik untuk setiap kolom
for kolom in data_train.columns:
    unique_values = data_train[kolom].unique()
    print(f"Nilai unik untuk kolom '{kolom}': {unique_values}")
```

```python
# Mengosongkan inputan yang berupa huruf
# Kolom target dikecualikan
target_column = 'Target'
for column in data_train.columns:
    if column != target_column:
        data_train[column] = data_train[column].apply(lambda x: np.nan if isinstance(x, str) and any(c.isalpha() for c in x) else x)
```

```python
# Melihat nilai unik untuk setiap kolom
for kolom in data_train.columns:
    unique_values = data_train[kolom].unique()
    print(f"Nilai unik untuk kolom '{kolom}': {unique_values}")
```

```python
data_train.info()
```

```python
data_train.isna().sum()
```

Jumlah data yang kosong bertambah

### Mengubah Tipe Data

```python
data_train.info()
```

```python
# List kolom-kolom yang ingin diubah menjadi tipe data numerik
kolom_numerik = [
    'Marital status', 'Application mode', 'Application order', 'Course',
    'Daytime/evening attendance\t', 'Previous qualification',
    'Previous qualification (grade)', 'Nacionality', "Mother's qualification",
    "Father's qualification", "Mother's occupation", "Father's occupation",
    'Admission grade', 'Displaced', 'Educational special needs', 'Debtor',
    'Tuition fees up to date', 'Gender', 'Scholarship holder',
    'Age at enrollment', 'International', 'Curricular units 1st sem (credited)',
    'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)', 'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (without evaluations)', 'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)', 'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (without evaluations)', 'Unemployment rate',
    'Inflation rate', 'GDP'
]


# Ubah tipe data kolom-kolom menjadi tipe data numerik
data_train[kolom_numerik] = data_train[kolom_numerik].apply(pd.to_numeric, errors='coerce')


# Periksa tipe data kolom-kolom setelah perubahan
print(data_train.dtypes)
```

### Imputasi Nilai

```python
# Salin data_train menjadi data_train_copy
data_train_copy = data_train.copy()


# Lakukan imputasi dengan mean
data_train_copy[kolom_numerik] = data_train_copy[kolom_numerik].fillna(data_train[kolom_numerik].mean())

# Periksa apakah imputasi berhasil
print(data_train_copy.isnull().sum())
```

```python
data_train_copy.head()
```

### Encoder

```python
# Inisialisasi LabelEncoder
label_encoder = LabelEncoder()

# Encode kolom target
data_train_copy['Target'] = label_encoder.fit_transform(data_train_copy['Target'])

# Lihat nilai-nilai yang diencode
print("Nilai-nilai yang diencode:", label_encoder.classes_)
```

## Fitting

```python
from sklearn.ensemble import ExtraTreesClassifier

# Inisialisasi model ExtraTreesClassifier
extra_trees_model = ExtraTreesClassifier(bootstrap=False, ccp_alpha=0.0, class_weight=None,
                     criterion='gini', max_depth=None, max_features='sqrt',
                     max_leaf_nodes=None, max_samples=None,
                     min_impurity_decrease=0.0, min_samples_leaf=1,
                     min_samples_split=2, min_weight_fraction_leaf=0.0,
                     n_estimators=100, n_jobs=-1,
                     oob_score=False, random_state=123, verbose=0,
                     warm_start=False)

# Latih model pada data train
extra_trees_model.fit(data_train_copy.drop(columns=['Target']), data_train_copy['Target'])

# Lakukan prediksi pada data test
predictions = extra_trees_model.predict(data_test_copy)
```

# File Submission

```python
# Tambahkan kolom prediksi ke dalam data_test_copy
data_test_copy['Target'] = predictions

# Tampilkan DataFrame data_test_copy dengan kolom prediksi
print(data_test_copy)
```
